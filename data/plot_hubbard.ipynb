```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'double_occupancy_results_1.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print("Data Preview:")
print(data.head())

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data['U/t'], data['Double Occupancy'], marker='o', linestyle='-', label=r"$\langle n \rangle$ = 1")

# Add labels, title, and legend
plt.title(r"1D Hubbard Model", fontsize=25)
plt.xlabel(r"Correlation strength $U/t$", fontsize=25)
plt.ylabel(r"Double occupancy $\langle n_{i\uparrow} n_{i\downarrow} \rangle$", fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.grid()
plt.legend(fontsize=20)

# Show the plot
plt.show()
```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths for each target occupancy
file_paths = {
    r"$\langle n \rangle$ = 1": 'double_occupancy_results_1.csv',
    r"$\langle n \rangle$ = 2/3": 'double_occupancy_results_0.6666666666666666.csv',
    r"$\langle n \rangle$ = 1/3": 'double_occupancy_results_0.3333333333333333.csv'
}

# Create a plot
plt.figure(figsize=(10, 6))

# Read and plot data for each target occupancy
for label, file_path in file_paths.items():
    data = pd.read_csv(file_path)
    plt.plot(data['U/t'], data['Double Occupancy'], marker='o', linestyle='-', label=label)

# Add labels, title, and legend
plt.title(r"1D Hubbard Model", fontsize=25)
plt.xlabel(r"Correlation strength $U/t$", fontsize=25)
plt.ylabel(r"Double occupancy $\langle n_{i\uparrow} n_{i\downarrow} \rangle$", fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.grid()
plt.legend(fontsize=20)

# Show the plot
plt.show()
```
