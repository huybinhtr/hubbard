```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = 'double_occupancy_results.csv'
data = pd.read_csv(file_path)


print("Data Preview:")
print(data.head())


plt.figure(figsize=(10, 6))
plt.plot(data['U/t'], data['Double Occupancy'], marker='o', linestyle='-', label=r"$\langle n \rangle$ = 1")


plt.title(r"1D Hubbard Model", fontsize=25)
plt.xlabel(r"Correlation strength $U/t$", fontsize=25)
plt.ylabel(r"Double occupancy $\langle n_{i\uparrow} n_{i\downarrow} \rangle$", fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.grid()
plt.legend(fontsize=20)


plt.show()
```
